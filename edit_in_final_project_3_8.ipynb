```python
project=pd.read_excel('/content/final_project.xlsx')
```

```python
project.columns
```

```python
# Calculate Shipping Delay
project["Shipping Delay"] = project["real day of shipping"] - project["scheduled days of shipment"]

# Add a column to describe the shipping status
project["Shipping Status"] = project["Shipping Delay"].apply(
    lambda x: "Shipped Earlier" if x < 0 else ("On Time" if x == 0 else "Shipped Late")
)

# Add a column to show the absolute number of days early or late
project["Shipping Delay Description"] = project["Shipping Delay"].apply(
    lambda x: f"{abs(x)} day(s) early" if x < 0 else ("On Time" if x == 0 else f"{x} day(s) late")
)

# Display the updated DataFrame
print(project[["real day of shipping", "scheduled days of shipment", "Shipping Delay", "Shipping Status", "Shipping Delay Description"]].head())
```

```python
project[["real day of shipping", "scheduled days of shipment", "Shipping Delay", "Shipping Status", "Shipping Delay Description"]]
```

```python
project['Order Status']=project['Order Status'].str.replace('_',' ')
```

```python
project['Order Status']
```

```python
project.drop(columns=['Shipping Delay','Shipping Status'],inplace=True)
```

```python
project = project.sort_values(by=['Customer Id','order date (DateOrders)' ])
project['Cumulative_Profit_by_Customer'] = project.groupby('Customer Id')['Order Profit Per Order'].cumsum()
```

```python
project['Cumulative_Profit_by_Customer']
```

```python


customer_order_counts = project['Customer Id'].value_counts().reset_index()
customer_order_counts.columns = ['Customer Id', 'Total Orders']

# Merge the new column with the original DataFrame
project = project.merge(customer_order_counts, on='Customer Id', how='left')
def classify_loyalty(orders):
    """Assigns a loyalty level based on the number of orders."""
    if orders >= 3:
        return 'High Loyalty'  # Frequent buyers
    elif orders >= 2:
        return 'Moderate Loyalty'  # Occasional buyers
    else:
        return 'Low Loyalty'  # One-time buyers

# Apply function to create loyalty column
project['Customer Loyalty'] = project['Total Orders'].apply(classify_loyalty)
```
